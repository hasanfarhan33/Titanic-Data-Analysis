```python
sns.pairplot(data = titanic_df, hue = "Sex")
```

```python
# Converting gender into a numerical data 
gender_dummies = pd.get_dummies(titanic_df["Sex"])
# gender_dummies
titanic_df = pd.concat([titanic_df, gender_dummies], axis = 1)
```

```python
titanic_df.head()
```

```python
# Dropping the unnecessary columns
titanic_df.drop(['Sex', 'female', 'PassengerId', 'Embarked', 'Name', 'Ticket'], axis = 1, inplace=True)
```

```python
titanic_df.head()
```

```python
from sklearn.model_selection import train_test_split 
y = titanic_df["Survived"]
X = titanic_df[["Pclass", "Age", "SibSp", "Parch", "Fare", "male"]]
```

```python
y 
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression 
logisticRegression = LogisticRegression()
```

```python
logisticRegression.fit(X_train, y_train)
```

```python
logisticRegression_predictions = logisticRegression.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, logisticRegression_predictions))
```

```python
print(classification_report(y_test, logisticRegression_predictions))
```
